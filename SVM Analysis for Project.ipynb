```python
# Define the sample size for balancing the data. 
n_rows_to_sample = 1000
```

```python
# Balance the data.
no_show_raw_df = raw_df[raw_df['No-show']=='Yes']
print(no_show_raw_df.shape)
no_show_raw_df = no_show_raw_df.sample(n = n_rows_to_sample, random_state = 0)
print(no_show_raw_df.shape)
```

```python
# Balance the data.
show_raw_df = raw_df[raw_df['No-show']=='No']
print(show_raw_df.shape)
show_raw_df = show_raw_df.sample(n = n_rows_to_sample, random_state = 0)
print(show_raw_df.shape)
```

```python
# Balance the data.
reduced_raw_df = no_show_raw_df.append(show_raw_df)
reduced_raw_df.shape
```

```python
# Show the first two rows.
reduced_raw_df.head(2)
```

```python
# Replace 'Yes' with the value 0 and 'No' with the value 1. 

reduced_raw_df.replace(['Yes', 'No'], value=[0, 1])
```

```python
reduced_raw_df.head(50)
```

```python
# Define target variables.
target = reduced_raw_df['No-show']
target_names = ['Yes', 'No']
```

```python
# Turn the strings into numbers.
reduced_raw_df['No-show'] = reduced_raw_df['No-show'].astype('category')
reduced_raw_df['No-show'] = reduced_raw_df['No-show'].cat.codes
```

```python
reduced_raw_df['PatientId'] = reduced_raw_df['PatientId'].astype('category')
reduced_raw_df['PatientId'] = reduced_raw_df['PatientId'].cat.codes
```

```python
reduced_raw_df['Gender'] = reduced_raw_df['Gender'].astype('category')
reduced_raw_df['Gender'] = reduced_raw_df['Gender'].cat.codes
```

```python
reduced_raw_df['ScheduledDay'] = reduced_raw_df['ScheduledDay'].astype('category')
reduced_raw_df['ScheduledDay'] = reduced_raw_df['ScheduledDay'].cat.codes
```

```python
reduced_raw_df['AppointmentDay'] = reduced_raw_df['AppointmentDay'].astype('category')
reduced_raw_df['AppointmentDay'] = reduced_raw_df['AppointmentDay'].cat.codes
```

```python
reduced_raw_df['Neighbourhood'] = reduced_raw_df['Neighbourhood'].astype('category')
reduced_raw_df['Neighbourhood'] = reduced_raw_df['Neighbourhood'].cat.codes
```

```python
# Show first fifty rows with data as integers. 
reduced_raw_df.head(50)
```

```python
reduced_raw_df['No-show'].unique()
```

```python
# important train test split

from sklearn.model_selection import train_test_split
y=reduced_raw_df['No-show']
X=reduced_raw_df.drop(['No-show'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC

model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report

predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
# Initialize SVM classifier
from sklearn import svm
clf_estimator_instance = svm.SVC(kernel='linear')
```

```python
# Fit data
clf_estimator_instance = clf_estimator_instance.fit(X_train, y_train)
```

```python
# Get the support vector indices with the estimator instance (clf)
svm_indices = clf_estimator_instance.support_
print(svm_indices)
```
