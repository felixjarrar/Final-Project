```python
n_rows_to_sample = 1000
```

```python
no_show_raw_df = raw_df[raw_df['No-show']=='Yes']
print(no_show_raw_df.shape)
no_show_raw_df = no_show_raw_df.sample(n = n_rows_to_sample, random_state = 0)
print(no_show_raw_df.shape)
```

```python
show_raw_df = raw_df[raw_df['No-show']=='No']
print(show_raw_df.shape)
show_raw_df = show_raw_df.sample(n = n_rows_to_sample, random_state = 0)
print(show_raw_df.shape)
```

```python
reduced_raw_df = no_show_raw_df.append(show_raw_df)
reduced_raw_df.shape
```

```python
reduced_raw_df.head(2)
```

```python
reduced_raw_df.replace(['Yes', 'No'], value=[0, 1])
```

```python
reduced_raw_df.head(50)
```

```python
target = reduced_raw_df['No-show']
target_names = ['Yes', 'No']
```

```python
reduced_raw_df['No-show'] = reduced_raw_df['No-show'].astype('category')
reduced_raw_df['No-show'] = reduced_raw_df['No-show'].cat.codes
```

```python
reduced_raw_df['PatientId'] = reduced_raw_df['PatientId'].astype('category')
reduced_raw_df['PatientId'] = reduced_raw_df['PatientId'].cat.codes
```

```python
reduced_raw_df['Gender'] = reduced_raw_df['Gender'].astype('category')
reduced_raw_df['Gender'] = reduced_raw_df['Gender'].cat.codes
```

```python
reduced_raw_df['ScheduledDay'] = reduced_raw_df['ScheduledDay'].astype('category')
reduced_raw_df['ScheduledDay'] = reduced_raw_df['ScheduledDay'].cat.codes
```

```python
reduced_raw_df['AppointmentDay'] = reduced_raw_df['AppointmentDay'].astype('category')
reduced_raw_df['AppointmentDay'] = reduced_raw_df['AppointmentDay'].cat.codes
```

```python
reduced_raw_df['Neighbourhood'] = reduced_raw_df['Neighbourhood'].astype('category')
reduced_raw_df['Neighbourhood'] = reduced_raw_df['Neighbourhood'].cat.codes
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X, y = reduced_raw_df.drop(['No-show'],axis=1), reduced_raw_df['No-show'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
X_train, y_train = np.array(X_train), np.array(y_train) 
#kf = KFold(n_splits=5, shuffle=True, random_state = 0)
```

```python
from sklearn.metrics import classification_report

logmodel = LogisticRegression(max_iter=1000)
# Train model
logmodel.fit(X_train,y_train)
# Get predictions
log_pred = logmodel.predict(X_test)
print(classification_report(y_test,log_pred))
```

```python
#class_weight='balanced'
logmodel = LogisticRegression(max_iter=1000, class_weight= 'balanced')
# Train model
logmodel.fit(X_train,y_train)
# Get predictions
log_pred = logmodel.predict(X_test)
print(classification_report(y_test,log_pred))
```
